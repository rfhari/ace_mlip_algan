```python
from ase.io import read, write

aln_unitcell = read("aln_unitcell.poscar", format='vasp')
gan_unitcell = read("gan_unitcell.poscar", format='vasp')

for fraction_al in [0.1, 0.3, 0.5, 0.7, 0.9]: # 0.0, 0.001, 0.01, 0.99, 0.999, 1.0
    fraction_ga = 1 - fraction_al

    averaged_atoms = aln_unitcell.copy()
    
    averaged_atoms.positions = aln_unitcell.positions * fraction_al + gan_unitcell.positions * fraction_ga
    
    new_cell = aln_unitcell.get_cell() * fraction_al + gan_unitcell.get_cell() * fraction_ga
    averaged_atoms.set_cell(new_cell)

    outfilename = f"averaged_{fraction_al:.1f}.poscar"
    write(outfilename, averaged_atoms, format='vasp')
    
    print(f"difference with AlN {averaged_atoms.positions - aln_unitcell.positions} for {fraction_al}Al")
    print(f"difference with GaN {averaged_atoms.positions - gan_unitcell.positions} for {fraction_al}Al")
```
