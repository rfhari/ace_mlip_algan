```python
def is_zero_matrix(matrix):
    return all(element == 0 for row in matrix for element in row)

def process_files_fc2_almabte(file, start_atom, neighbors):
    matrices = []
    num_zeros_matrix = 0
    lines_to_skip = 1 + 576 * 4 * (start_atom-1)
    with open(file, 'r') as f1: 
        for _ in range(lines_to_skip):  
            f1.readline()
        line = f1.readline()
        print("values:", list(map(float, line.strip().split())))
        while len(matrices) < neighbors:
            matrix = []
            for _ in range(3):  
                line = f1.readline()
                if not line.strip():
                    break
                matrix.append(list(map(float, line.strip().split())))
            if len(matrix) == 3:
                if is_zero_matrix(matrix):
                    num_zeros_matrix+=1
                matrices.append(matrix)
            if len(matrices)%neighbors==0:
                break
                print("finished reading")
            else:               
                for _ in range(1):  
                    f1.readline()
            # if len(matrices) >= neighbors:
            #     break
    return matrices, num_zeros_matrix

for ii in range(1, 101):
    fc2_gan_almabte, num_zeros_matrix = process_files_fc2_almabte("FORCE_CONSTANTS_2ND_GaN_TDEP", start_atom=ii, neighbors=576)
    # fc2_gan_almabte, num_zeros_matrix = process_files_fc2_almabte("FORCE_CONSTANTS_ALAMODE_GaN", start_atom=ii, neighbors=576)
    print("total and zero block:", len(fc2_gan_almabte), num_zeros_matrix)
```
